```python
training_set = pd.concat([training["Label"],df], axis = 1)
```

```python
nspam = training_set[training_set["Label"]=="spam"].sum(axis = 1).sum()
```

```python
nham = training_set[training_set["Label"]=="ham"].sum(axis = 1).sum()
```

```python
# P(Spam|New_message) = P(Spam)*nword + alpha/(nspam + alpha*n_v)*nword + alpha/(nspam + alpha*n_v)
# P(ham|New_message) = P(ham)*nword + alpha/(nham + alpha*n_v)*nword + alpha/(nspam + alpha*n_v)
```

```python
spam = training_set[training_set["Label"]=="spam"]
ham = training_set[training_set["Label"]=="ham"]
```

```python
spam.head()
```

```python
spam_parameter = {i:0  for i in vocabulary}
ham_parameter = {i:0  for i in vocabulary}
```

```python
for i in vocabulary:
    nword = spam[i].sum() 
    p_word_given_spam = (nword + 1)/nspam+1*len(vocabulary)
    spam_parameter[i] = p_word_given_spam
    
    
```

```python
for i in vocabulary:
    nword = ham[i].sum() 
    p_word_given_ham = (nword + 1)/nham+1*len(vocabulary)
    ham_parameter[i] = p_word_given_ham
```

```python
import re
def classify(new_message):
    
    msg = re.sub(r"\W"," ", new_message).lower().split()
    P_Spam_new_message = p_spam
    P_ham_new_message = p_ham
    
    
    for word in msg:
        if word in spam_parameter:
            P_Spam_new_message *= spam_parameter[word]
            P_ham_new_message *= ham_parameter[word]
        else:
            return "Human will classify"
    
    if P_Spam_new_message > P_ham_new_message:
        return "spam"
    
    elif P_Spam_new_message < P_ham_new_message:
        return "ham"
        
    
```

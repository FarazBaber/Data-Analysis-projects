```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stories = pd.read_csv("sel_hn_stories.csv", header = None, 
                      names = ["sub_time", "upvote", "url", "headlines"])
```

```python
stories.head()
```

```python
stories.dropna(axis = 0, inplace = True)
stories.reset_index(inplace=True)
```

```python
punctuation = [",", ":", ";", ".", "'", '"', "’", "?", "/", "-", "+", "&", "(", ")"]
```

```python
headlines = stories["headlines"]
```

```python
headlines.head()
```

```python
headlines = headlines.str.lower()
```

```python
headlines.head()
```

```python
"Muhammad Waqas".split()
```

```python
tokens = []

for i in headlines:
    tokens.append(i.split())
```

```python
print(tokens)
```

```python
# nested loop

lst = []
for i in tokens:
    clean = []
    for j in i:
        if j not in punctuation:
            clean.append(j)
    lst.append(clean)

print(lst)
```

```python
import string
```

```python
dir(string)
```

```python
unique = []

for i in lst:
    for j in i:
        unique.append(j)

print(unique[:10])
```

```python
for i in punctuation:
    print(i in unique)
```

```python
nlp = pd.DataFrame(data = 0, index = headlines.index, columns = unique)
```

```python
nlp.head()
```

```python
for i, lst1 in enumerate(lst):
    for token in lst1:
        nlp.loc[i, token] += 1
```

```python
count = nlp.sum(axis=0)
count
```
